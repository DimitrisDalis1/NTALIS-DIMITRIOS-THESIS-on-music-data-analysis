Imports

```python
import pandas as pd
import numpy as np
from sklearn.manifold import MDS
import matplotlib.pyplot as plt
from scipy.spatial.distance import cityblock
from scipy.spatial.distance import cdist
from sklearn.metrics.pairwise import cosine_distances
from scipy.spatial.distance import jensenshannon
```

```python
#Load data
data = pd.read_csv('extracted_data.csv')

# Separate numerical data: from the beginning to the 4th column before the end
numerical_data = data.iloc[:, :-4]

# Separate artist category: the 3rd column from the end
artist_category = data.iloc[:, -3]

X = data.values

distance_matrix = cdist(numerical_data.values, numerical_data.values, metric='cityblock')

mds = MDS(n_components= 2, random_state=0, dissimilarity='precomputed')
X_2d = mds.fit_transform(distance_matrix)

unique_artists = artist_category.unique()
color_map = {artist: plt.cm.tab10(i) for i, artist in enumerate(unique_artists)}

# Plot the data with different colors for each artist
plt.figure(figsize=(10, 8))
for artist in unique_artists:
    artist_data = X_2d[artist_category == artist]
    plt.scatter(artist_data[:, 0], artist_data[:, 1], color=color_map[artist], label=artist)

plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.title('MDS Visualization by Artist')
plt.legend()
plt.show()
```
